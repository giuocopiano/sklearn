## How is the Happiness Score calculated? What is its relation with the other indicators? Can it be derived from the others?

The following is a predictive analysis of the happiness score using the 2015 data. 

```python
target = data2015['Happiness Score']
features = data2015.drop(['Country', 'Region', 'Happiness Score', 'Happiness Rank'], axis = 1)
```

```python
target.head()
```

```python
features.head()
```

```python
import sklearn
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(2)
poly.fit_transform(X_train)
poly.transform(X_test);
```

```python
from sklearn.linear_model import Ridge
clf = Ridge(alpha = 1.0)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, predictions)
```

Yes, the Happiness Score is readily derivable from the rest of the indicators with a high degree of accuracy.
