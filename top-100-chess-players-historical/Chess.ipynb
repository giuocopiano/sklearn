## Kasparov and his contemporaries

```python
kasparov = data.loc[data['name'] == 'Kasparov, Garry']
anand = data.loc[data['name'] == 'Anand, Viswanathan']
kramnik = data.loc[data['name'] == 'Kramnik, Vladimir']
topalov = data.loc[data['name'] == 'Topalov, Veselin']
adams = data.loc[data['name'] == 'Adams, Michael']
moro = data.loc[data['name'] == 'Morozevich, Alexander']
```

```python
plt.figure()
plt.plot(range(22), kasparov['rating'][:22], label = 'Kasparov')

plt.plot(range(22), anand['rating'][:22], label = 'Anand')

plt.plot(range(22), kramnik['rating'][:22], label = 'Kramnik')

plt.plot(range(22), topalov['rating'][:22], label = 'Topalov')

plt.plot(range(22), moro['rating'][:22], label = 'Moro')

plt.plot(range(22), adams['rating'][:22], label = 'Adams')

plt.legend(prop={'size':7})
```
